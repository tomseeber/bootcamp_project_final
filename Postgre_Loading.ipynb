```python
username = 'postgres'
password = 'postgres'
database_name = 'housing_data_db'
host = 'localhost'  
port = '5432'  

# Create the connection string for SQLAlchemy
connection_string = f'postgresql://{username}:{password}@{host}:{port}/{database_name}'

# Create the database engine
engine = create_engine(connection_string)

# Read the re-cleaned data into a DataFrame
re_cleaned_df = pd.read_csv('Data/re_cleaned_housing_data.csv')

# Write records stored in the DataFrame to the SQL database
re_cleaned_df.to_sql('housing', engine, if_exists='replace', index=False)

```

```python
# Sample code to perform a query check
test_df = pd.read_sql_query('SELECT * FROM housing', engine)
print(f"Number of rows in the database table: {len(test_df)}")
test_df.head()
```
